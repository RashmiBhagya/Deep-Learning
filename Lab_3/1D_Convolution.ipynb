```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal = np.array([1, 2, 1, 2, 1, 1, 1, 0.5, 0.5, 0.5, 1, 1, 1])
```

```python
plt.stem(signal)
plt.show()
```

```python
oneD_filter = np.array([-2, 1])
```

```python
plt.stem(oneD_filter)
plt.show()
```

```python
result = np.convolve(signal, oneD_filter, 'valid')
```

```python
plt.stem(result)
plt.show()
```
