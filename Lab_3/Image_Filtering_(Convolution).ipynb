```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

### Load the image using OpenCV

```python
im = cv2.imread('/content/sample_data/lenna.png') #change the path as needed
```

```python
if im is not None:
    # Get the shape of the image (height, width, number of channels)
    height, width, channels = im.shape
    print(f"Image shape: {height} x {width} x {channels}")
else:
    print("Image not loaded successfully. Please check the file path.")
```

### Convert the RGB image into a grayscale image

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

### Define vertical and horizontal filter kernels

```python
v_kernel = np.array([
    [-1, 0, 2],
    [-1, 0, 2],
    [-1, 0, 2]
])
```

```python
h_kernel = np.array([
    [-15, -15, -15],
    [0, 0, 0],
    [15,15, 15]
])
```

### Do the filtering

```python
v_image = cv2.filter2D(im, -3, v_kernel)
h_image = cv2.filter2D(im, -3, h_kernel)
```

### Display results

```python
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='Blues')
plt.title('Horizontal Image')
```
